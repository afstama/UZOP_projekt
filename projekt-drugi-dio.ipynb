### Rukovanje nedostajućim podacima

```python
df.isna().sum()
```

S obzirom na to da postoji puno nedostajućih podataka, ovdje ćemo morati raditi dosta izabcivanja/izmjena. U slučaju nedostajućih podataka primijenit ćemo jednu od tri opcije:
1. Izbacivanje značajke
2. Izbacivanje zapisa
3. Zamjena nedostajućih vrijednosti sredjom vrijednosti.

Iako nedostajuće vrijednosti ciljne značajke iznose čak 40.44% svih podataka, svejedno ih moramo ukloniti iz skupa podataka. Ne možemo maknuti cijelu značajku jer je ciljna, ali ne možemo ni zamijeniti nedostajuće vrijednosti srednjom vrijednosti jer tako nećemo dobiti kasnije točne prediktivne modele zbog "lažnih" podataka. Stoga ćemo odmah ukloniti zapise s nedostajućim vrijednostima značajki _LNH_.

```python
df = df.loc[df['LNH'].notnull(), :]
print(len(df))
df.isna().sum()
```

Izabcivanjem zapisa s nedostajućim vrijednostima _LNH_ dobili smo skup podataka koji ima 1072 zapisa, što čini 59.56% originalnog skupa podataka. Sad ćemo izračunati postotke preostalih nedostajućih vrijednosti po značajkama:

- SEC &ndash; 18.66%
- Age(years) &ndash; 0.28%
- Height(cm) &ndash; 11.94%
- Bgroup &ndash; 34.24%
- Iwt(kg) &ndash; 1.12%
- Fwt(kg) &ndash; 1.21%
- IBP_sys &ndash; 0.37%
- IBP_dias &ndash; 0.37%
- FBP_sys &ndash; 0.56%
- FBP_dias &ndash; 0.56%
- IHb(gm%) &ndash; 9.61%
- FHb(gm%) &ndash; 9.52%
- BS(RBS) &ndash; 49.53%
- Term/Preterm &ndash; 12.41%
- Sex &ndash; 1.03%

Značajke koje imaju veliki broj nedostajućih vrijednosti te kategoričke značajke čiju srednju vrijednost ne možemo izračunati ćemo potpuno ukloniti iz skupa podataka: _SEC_, _Bgroup_, _BS(RBS)_ i _Term/Preterm_.

U slučaju značajki s manje od 2% nedostajućih vrijednosti uklonit ćemo zapise koji sadrže nedostajuće vrijednosti: _Age(years)_, 'IWt(kg)_, _FWt(kg)_, _IBP\_sys_, _IBP\_dias_, _FPB\_sys_, _FBP\_dias_ i _Sex_.

U slučaju preostalih značajki nedostajuće vrijednosti zamijenit ćemo srednjom vrijednosti značajke: _Height(cm)_, _IHb(gm%)_ i _FHb(gm%)_.

```python
df_tmp = df.copy()
# izbacivanje značajki
df_tmp.drop(['SEC', 'Bgroup', 'BS(RBS)', 'Term/Preterm'], axis=1, inplace=True)
# zamjena srednjom vrijednosti
df_tmp.loc[df_tmp['Height(cm)'].isna(),'Height(cm)'] = df_tmp.loc[:,'Height(cm)'].mean()
df_tmp.loc[df_tmp['IHb(gm%)'].isna(),'IHb(gm%)'] = df_tmp.loc[:,'IHb(gm%)'].mean()
df_tmp.loc[df_tmp['FHb(gm%)'].isna(),'FHb(gm%)'] = df_tmp.loc[:,'FHb(gm%)'].mean()
# izbacivanje zapisa
df_tmp = df_tmp.dropna()
```

```python
df = df_tmp.copy()
df.describe()
```

```python
print(f'Broj zapisa = {len(df)}')
print(f'Broj značajki = {len(df.columns)}')
```

Nakon što smo se riješili nedostajućih vrijednosti, smanjili smo broj zapisa s 1800 na 1044, što je 58% svih podataka. Također smo smanjili broj značajki s 18 na 14.

### Provjera i pretvorba tipova podataka

```python
df.dtypes
```

Značajke _Sex_ i _LNH_ su kategoričke, pa ćemo ih pretvoriti u numeričke varijable.

```python
df['LNH'] = df['LNH'].astype('int')

df['Sex'] = df['Sex'].astype('category')
conversionValues = dict(enumerate(df['Sex'].cat.categories))
df['Sex'] = df['Sex'].cat.codes

df.dtypes
```

Za numeričke značajke prikazat ćemo srednju, centralnu, minimalnu i maksimalnu vrijednost, dok ćemo za kategoričke značajke ispisati učestalost pojavljivanja svake vrijednosti.

```python
categoricals = set(['Sex', 'LNH'])
numericals = list(set(df.columns) - categoricals)


# numeričke značajke
for col in numericals:
    print(f'\n{col} MEAN = {df[col].mean()}')
    print(f'{col} MEDIAN = {df[col].median()}')
    print(f'{col} MIN = {df[col].min()}')
    print(f'{col} MAX = {df[col].max()}')

# kategoričke značajke
for col in categoricals:
    print('\n', col, '\n', df[col].value_counts())
```

Što se tiče kategoričkih značajki, ciljna značajka _LNH_ ima samo dvije vrijednosti jednake 3 (koje odgovaraju prevelikoj težini), što čini 0.38% svih podataka. S obzirom na to da ćemo teško kasnije naučiti model da generalizira na temelju 4 zapisa, uklonit ćemo ih iz skupa podataka. Isto to ćemo učiti i s kategoričkom značajkom _Sex_, kojoj su dvije vrijednosti jednake 2, što čini 0.19% svih podataka.

```python
df = df[df['LNH']!=3]
df = df[df['Sex']!=2]
del conversionValues[2]
for col in categoricals:
    print('\n', col, '\n', df[col].value_counts())
```

Numeričke značajke normalizirat ćemo na vrijednosti između 0 i 1.

```python
df_vis = df.copy()
for col in numericals:
        df[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())
```

### Vizualizacija podataka

Kako bismo dobili detaljniji uvid u podatke, predstavit ćemo ih vizualno.
